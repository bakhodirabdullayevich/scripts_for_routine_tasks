```python
import pandas as pd

# Определяем словари для соответствия кода области и района
oblasti_dict = {
    '35': 'Qoraqalpog‘iston Respublikasi', '03': 'Andijon viloyati', '06': 'Buxoro viloyati',
    '08': 'Jizzax viloyati', '10': 'Qashqadaryo viloyati', '12': 'Navoiy viloyati',
    '14': 'Namangan viloyati', '26': 'Toshkent shahri', '18': 'Samarqand viloyati',
    '24': 'Sirdaryo viloyati', '22': 'Surxondaryo viloyati', '27': 'Toshkent viloyati',
    '30': 'Farg‘ona viloyati', '33': 'Xorazm viloyati'
}

rayon_dict = {
    '35': {'204': 'Amudaryo tumani', '207': 'Beruniy tumani', '209': 'Bo‘zatov tumani', '211': 'Qorao‘zak tumani', '212': 'Kegeyli tumani'},
    '03': {'202': 'Oltinko‘l tumani', '203': 'Andijon tumani', '206': 'Baliqchi tumani', '209': 'Bo‘ston tumani', '210': 'Buloqboshi tumani', '211': 'Jalaquduq tumani'},
    '06': {'204': 'Olot tumani', '207': 'Buxoro tumani', '212': 'Vobkent tumani', '215': 'G‘ijduvon tumani'}
}

# Открываем файл Excel (замените путь на реальный)
file_path = 'file.xlsx'
df = pd.read_excel(file_path)

# Функция для извлечения области и района на основе кода
def get_oblast_rayon(kod_objekta):
    kod_objekta = kod_objekta.strip('="')
    oblast_code = kod_objekta[2:4]
    rayon_code = kod_objekta[4:7]

    oblast = oblasti_dict.get(oblast_code, 'Неизвестная область')
    rayon = rayon_dict.get(oblast_code, {}).get(rayon_code, 'Неизвестный район')
    return oblast, rayon

# Применяем функцию к столбцу 'Код объекта' и добавляем новые столбцы 'Область' и 'Район'
df[['Область', 'Район']] = df['Код объекта'].apply(lambda x: pd.Series(get_oblast_rayon(x)))

# Сохраняем результат в новый Excel файл
df.to_excel('с_областью_и_районом.xlsx', index=False)

print("Готово! Результат сохранен в файле 'с_областью_и_районом.xlsx'")
```

```python
import pandas as pd

# Замените пути на реальные пути к вашим Excel файлам
file_path_1 = 'File-1.xlsx'
file_path_2 = 'File-2.xlsx'


# Открываем оба файла с принудительным преобразованием столбца "Код объекта" в текстовый тип
df1 = pd.read_excel(file_path_1, dtype={'Код объекта': str})
df2 = pd.read_excel(file_path_2, dtype={'Код объекта': str})

# Удаляем пробелы из названий столбцов
df1.columns = df1.columns.str.strip()
df2.columns = df2.columns.str.strip()

# Проверяем, есть ли столбец "Код объекта"
print("Столбцы в первом файле:", df1.columns)
print("Столбцы во втором файле:", df2.columns)

# Если столбца "Код объекта" нет, то добавим его с пустыми значениями
if 'Код объекта' not in df1.columns:
    raise KeyError("Столбец 'Код объекта' не найден в первом файле.")
if 'Код объекта' not in df2.columns:
    raise KeyError("Столбец 'Код объекта' не найден во втором файле.")

# Приводим столбцы "Код объекта" к единому формату: удаляем пробелы и любые скрытые символы
def clean_code_column(df, column_name):
    df[column_name] = df[column_name].astype(str)  # Преобразуем к строке
    df[column_name] = df[column_name].str.replace(r'\s+', '', regex=True)  # Удаляем пробелы
    df[column_name] = df[column_name].str.replace(r'[^\w]', '', regex=True)  # Убираем спецсимволы
    return df

df1 = clean_code_column(df1, 'Код объекта')
df2 = clean_code_column(df2, 'Код объекта')

# Выполняем объединение с индикатором, чтобы отслеживать совпадения и расхождения
merged_df = df1.merge(df2, on='Код объекта', how='outer', suffixes=('_file1', '_file2'), indicator=True)

# Сначала размещаем совпадающие строки, затем — несовпадающие
matching_df = merged_df[merged_df['_merge'] == 'both']
non_matching_df = merged_df[merged_df['_merge'] != 'both']

# Объединяем результат в один DataFrame, где совпадающие строки идут первыми
final_df = pd.concat([matching_df, non_matching_df]).reset_index(drop=True)

# Удаляем индикаторный столбец
final_df = final_df.drop(columns=['_merge'])

# Сохраняем результат в новый Excel файл
final_df.to_excel('1совпадающие_и_несовпадающие_данные.xlsx', index=False)

print("Готово! Результат сохранен в файле 'совпадающие_и_несовпадающие_данные.xlsx'")
```
